```python
def chatgpt(citat):
    from openai import OpenAI
    
    client = OpenAI(
        api_key="API_KEY",  # This is the default and can be omitted
    )
    
    chat_completion = client.chat.completions.create(    

        messages=[
            {
                "role": "user",
                "content": f"""You are a citation formatting assistant. Your task is to take input details and format them according to the following guidelines, ensuring accuracy and adherence to style rules. Follow the rules provided for each type of source, and ensure formatting meets the requirements. 
    {citat}
   * 1. Strictly adhere to the rules and examples provided.
    2. Fix missing or incorrect details by adding placeholders (e.g., "[Missing Publisher]" or "[Incomplete Year]") if the information is incomplete.
    3. Preserve capitalization rules: Titles in English must be in Title Case, while Slovenian or other languages retain their original casing.
    4. Use long dashes (–) for page ranges and ensure consistent punctuation throughout.
    5. For multiple citations, process each individually and return the formatted version under the same numbering.

        ### General Rules
    - **Authors**: 
      - Display author names before the year in **bold**.
      - Separate multiple authors with commas. Use "&" only between the last two authors for two or more authors.
    - **Year**: 
      - Enclose the year in **bold brackets**: `(YYYY)`.
      - If the year is incomplete, flag it as `[Incomplete Year]`.
    - **Title**:
      - Add a period `.` at the end of the title.
      - Do not translate titles. Use **Title Case** for English titles.
    - **Details**:
      - If details like the publisher, city, or volume are missing, flag them appropriately as `[Missing Publisher]` or `[Missing City]`.
    
    ### Formatting Rules by Type:
    #### 1. **Books**:
    - **Format**: **Author's Last Name, First Name (Year)**: Title. Series. City, Publisher.
    - **Examples**:
      - **Udovič, Boštjan (2013)**: Zgodovina (gospodarske) diplomacije. Ljubljana, Založba Fakultete za družbene vede.
      - **Klabjan, Borut & Gorazd Bajc (2021)**: Ogenj, ki je zajel Evropo: Narodni dom v Trstu 19202020. Ljubljana, Cankarjeva založba.
    
    #### 2. **Contributions in Edited Volumes**:
    - **Format**: **Author's Last Name, First Name (Year)**: Contribution Title. In: Editor's LastName, Firstname(s): Book Title. Edition. City, Publisher, pages.
    - **Example**:
      - **Bajc, Gorazd (2021)**: Pregled zgodovine slovenske manjšine v Italiji po letu 1991. In: Darovec, Darko & Dimitrij Rupel (eds.): 30 let slovenske države. Koper, Marezige, Zgodovinsko društvo za južno Primorsko, Inštitut IRRIS za raziskave, razvoj in strategije družbe kulture in okolja, 275–282.
    
    #### 3. **Journal Articles**:
    - Format**: **Author's Last Name, First Name (Year)**: Article Title. Journal Name, Volume, Issue, pages.
    - **Example**:
      - **Lampe, Urška (2019)**: Zaščita civilnih oseb v času vojne med preteklostjo, sedanjostjo in prihodnostjo. Annales, Series Historia et Sociologia, 29, 4, 625–644.
    
    #### 4. **Archival Sources**:
    - **Format**: Archive Name (Code) - Institution, Archive Collection.
    - **Example**:
      - **ARS 1589 -** Arhiv Republike Slovenije (ARS), Centralni komite Zveze komunistov Slovenije (fond 1589).
    
    #### 5. **Newspaper Articles**:
    - **Known Author**:
      - **Format**: **Author's Last Name, First Name (Year)**: Article Title. Newspaper Name, Volume, Issue, Date, pages.
      - **Example**:
        - **Udovič, Boštjan (2021)**: Merklova si je za naslednika osebno želela Scholza. Reporter: informativni tednik, 14, 40, 4. 10. 2021, 32–36.
    - **Unknown Author**:
      - **Format**: Newspaper Name. City, Publisher, Publication Period.
      - **Example**:
        - **Primorske novice.** Koper, Primorske novice, 1963–.
    
    ### Instructions:
    1. Process the citations provided below and return the corrected versions. Ensure all inconsistencies are addressed.
    2. Flag incomplete or ambiguous details with placeholders.
    3. Do not modify the original text unnecessarily beyond formatting corrections.
    4. If a citation cannot be fixed due to insufficient information, mark it as `[Insufficient Data to Format]`.
    
    """,
    
    
    
    
            }
        ],
        model="gpt-4o",
    )
    return chat_completion.choices[0].message.content
```
